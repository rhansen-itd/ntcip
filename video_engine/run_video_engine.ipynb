```python
import sys
import os
from pathlib import Path

# 1. Figure out exactly where we are right now
current_cwd = Path.cwd().resolve()

# 2. Dynamically determine the paths based on where VSCode started the notebook
if current_cwd.name == 'video_engine':
    # Notebook is running inside the video_engine folder
    ntcip_root = current_cwd.parent
    video_engine_dir = current_cwd
else:
    # Notebook is running from the ntcip workspace root
    ntcip_root = current_cwd
    video_engine_dir = current_cwd / 'video_engine'

# 3. Inject both paths into the system path so Python can find EVERYTHING
if str(ntcip_root) not in sys.path:
    sys.path.insert(0, str(ntcip_root))        # This allows: import ntcip_monitor
    
if str(video_engine_dir) not in sys.path:
    sys.path.insert(0, str(video_engine_dir))  # This allows: import system_runner

# 4. Verify it worked!
print(f"NTCIP Root added to path: {ntcip_root}")
print(f"Video Engine added to path: {video_engine_dir}")

# 5. Do your imports
from system_runner import SystemRunner
print("\nImports successful!")
```

```python
# 2. Ensure output directories exist
os.makedirs("./trigger_queue", exist_ok=True)
os.makedirs("./completed_videos", exist_ok=True)

print("üöÄ Initializing System Runner...")
runner = SystemRunner(
    intersection_id="201",
    config_path="intersections.json",
    trigger_dir="./trigger_queue",
    output_dir="./completed_videos",
    min_free_disk_mb=500.0,
    max_concurrent_writers=2,
)

try:
    print("üü¢ System is LIVE. Listening for NTCIP events and buffering video.")
    print("‚èπÔ∏è  Press the Jupyter 'Stop' button (Interrupt Kernel) to shut down gracefully.")
    
    # This call blocks the cell. The background threads will do all the work.
    runner.run()

except KeyboardInterrupt:
    print("\n‚ö†Ô∏è Jupyter Interrupt received! Triggering graceful shutdown...")

finally:
    # This guarantees your Video Buffers safely write the .mp4 files before closing
    runner.shutdown()
    print("üõë System fully stopped.")
```
